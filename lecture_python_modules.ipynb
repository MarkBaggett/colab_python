# Python Modules

When the Python interpreter starts you get access to your keywords and one hundred or so preloaded variables. Python also comes with other loadable modules. These loadable modules also define new variables that you can use in your programs. Most of those variables have code in them you can call and execute.

Here are the additional "builtin" modules:

https://docs.python.org/3/py-modindex.html

# Importing Modules

You add the functions that are in these modules using keyword "import". 


```python
import os
os.getcwd()   #Notice this has the module name and the function in the module
```

You will see a second way to import things using the keyword "from".  With this syntax you specify an item that is in the module, then when you call it you to not provide the module name 

```python
from os import getcwd
getcwd()      #Notice that this one does not have the module name.
```
