# Python Numeric Types

Python has several Numeric Type Variables.  Integers, Floats and Complex Numbers

Integers (int): Integers are whole numbers, such as 1, 2, 3, 4, etc. They have no decimal point.

Floating-point numbers (float): Floating-point numbers are numbers with a decimal point, such as 3.14, 2.0, -1.5, etc.

Complex numbers (complex): Complex numbers have real and imaginary parts.  We won't use these in this class.


```
x = 0b101

```
### Octal notation
```
x = 0o12
print(x)  # Output: 10
```

````python
# Initialize an integer variable
x = 10

# Basic arithmetic operations
print(x + 5)  # Addition: 15
print(x - 3)  # Subtraction: 7
print(x * 2)  # Multiplication: 20
print(x / 3)  # Division: 3.33333
print(x // 3) # Integer division: 3
print(x % 3)  # Modulo: 1
print(x ** 2) # Exponentiation: 100

# Bitwise operations
print(x & 7)  # Bitwise AND: 2
print(x | 7)  # Bitwise OR: 15
print(x ^ 7)  # Bitwise XOR: 13
print(~x)     # Bitwise NOT: -11
print(x << 2) # Bitwise left shift: 40
print(x >> 2) # Bitwise right shift: 2

# Comparison operations
print(x == 10) # Equality: True
print(x != 10) # Inequality: False
print(x > 5)   # Greater than: True
print(x < 5)   # Less than: False
print(x >= 10) # Greater than or equal to: True
print(x <= 5)  # Less than or equal to: False

# Other methods
print(x.bit_length()) # Number of bits needed to represent x in binary: 4
print(x.to_bytes(2, byteorder='big')) # Integer x as bytes in big-endian byte order: b'\x00\n'
````

````python
# Initialize a float variable
x = 3.14159

# Basic arithmetic operations
print(x + 2.71828)  # Addition: 5.85987
print(x - 1.0)      # Subtraction: 2.14159
print(x * 2.0)      # Multiplication: 6.28318
print(x / 2.0)      # Division: 1.570795
print(x // 2.0)     # Integer division: 1.0
print(x % 2.0)      # Modulo: 1.14159
print(x ** 2.0)     # Exponentiation: 9.8695877281

# Comparison operations
print(x == 3.14159)     # Equality: True
print(x != 3.14159)     # Inequality: False
print(x > 3.0)          # Greater than: True
print(x < 3.0)          # Less than: False
print(x >= 3.14159)     # Greater than or equal to: True
print(x <= 2.0)         # Less than or equal to: False

# Other methods
print(x.is_integer())   # False
print(x.hex())          # Return hexadecimal string representation of x: '0x1.921f9f01b866ep+1'
print(x.as_integer_ratio())  # Return (numerator, denominator) pair of x as ints: (3537118815677477, 1125899906842624)
print(float.fromhex('0x1.921f9f01b866ep+1'))  # Convert hexadecimal string to float: 3.14159
````

### Accuracy beyond 15 decimal places.

Python has a module called Decimal that can offer you accuracy beyond 15 decimal points. When using the module it is best to provide the values as strings and not as floats.  

````python
from decimal import Decimal
x = Decimal("0.1")
y = Decimal("0.2")
x + y
````

If you provide the values as floats the decimal module will blindly accept the inaccuracies created by floats in its calculations.

````python
x = Decimal(0.1)
y = Decimal(0.2)
x + y
````
